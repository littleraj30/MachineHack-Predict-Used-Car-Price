```python
def CrossVal(dataX,dataY,mode,cv=10):
    score=cross_val_score(mode,dataX , dataY, cv=cv, scoring=metric)
    print("cv score",np.mean(score))
    return(np.mean(score))
```

```python
def rmsle(y_pred, y_test) : 
    return(1-(np.sqrt(np.mean((np.log1p(y_pred) - np.log1p(y_test))**2))))
metric = make_scorer(rmsle)
```

```python
def stratified(model , data = train, label= train_label):
    cv_ans = pd.DataFrame(data = [], index = data.index)
    cv_list = []
    r2_score = []
    k = KFold(n_splits =5,shuffle =True, random_state = 38543)
    for train, cv in k.split(data):
        Xtrain = data.iloc[train,:]
        Ytrain = label.iloc[train]
        Xcv = data.iloc[cv,:]
        Ycv = label.iloc[cv]
        model.fit(Xtrain, Ytrain)
        score=rmsle(model.predict(Xcv),Ycv)
        score_r2=r2(model.predict(Xcv),Ycv)
        cv_list.append(score)
        r2_score.append(score_r2)
        
    print("Mean score: ",np.mean(cv_list),"\n Mean R2 :",np.mean(r2_score) , "\n MSE :" ,
          np.sqrt(mean_squared_error(model.predict(Xcv),Ycv)))
    sns.scatterplot(model.predict(Xcv), Ycv)
```

```python
def stacking(model , name,data = train, label= train_label, test = test):
    i=0
    cv_ans = pd.DataFrame({name+' Prediction' : 0}, index = data.index)
    test_ans = pd.DataFrame( data=[] ,index = test.index, columns = [i for i in range(5)])
    k = KFold(n_splits = 5 , shuffle = False, random_state = 554)
    for train, cv in k.split(data):
        Xtrain = data.iloc[train,:]
        Ytrain = label.iloc[train]
        Xcv = data.iloc[cv,:]
        Ycv = label.iloc[cv]
        
        model.fit(Xtrain, Ytrain)
        cv_ans.loc[cv,name+' Prediction' ] = model.predict(Xcv)
        test_ans.iloc[:,i] = model.predict(test)
        i=i+1
    return(cv_ans, pd.DataFrame({'mean_val'+name: test_ans.mean(axis=1)}))
```
